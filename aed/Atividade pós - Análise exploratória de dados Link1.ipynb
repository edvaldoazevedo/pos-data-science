```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('DIMIF-queima.csv', delimiter=';')# Importação do dataset com configuração do delimitador ;
```

```python
data
```

```python
data_by_governo=data['Coordenação Regional do ICMBio'].value_counts()#Resolução primeiro ponto onde pedia-se para apresentar o governo que mais criou UCs durantes os anos da pesquisa
#Sendo 41 UCs criadas pelo governo de SC
```

```python
data_by_governo # Apresentando resultado
```

```python
data_by_ctg_bio= data.groupby(['Bioma referencial','Categoria da UC: nomenclatura nacional']).count()
# Segundo ponto pedia-se para criar novo dataframe agrupando por categoria de UC e Bioma Referencial para isso foi utilizado o comando groupby por ser mais utilizado e agrupa de forma rápida
```

```python
data_by_ctg_bio['Área queimada em 2018'].plot(figsize=(20,8)) #Após a criação do novo dataframe pedia-se para criar um grafico que apresentasse resultados relacionados com os dois campos agrupados

#Foi utililizado o campo Área queimada em 2018 para apresentar a quantidade de área queimada em 2019 relacionado entre as categorias de UCs e Bioma
```

```python
data.rename(columns={'Área queimada em 2018':'2018','Área queimada em 2017':'2017','Área queimada em 2016':'2016','Área queimada em 2015':'2015','Área queimada em 2014':'2014','Área queimada em 2013':'2013','Área queimada em 2012':'2012'},inplace=True)
#Para o terceiro ponto se fez necessário renomear as colunas relacionadas as areas queimadas pelos anos, aqui utilizou o comando rename, passando como parametros as colunas a serem renomeadas e seus novos valores
```

```python
data_por_ano = data[['2012','2013','2014','2015','2016','2017','2018']].count() # A contagem de areas queimadas pelos anos da pesquisa
```

```python
data_por_ano # Resultado Terceiro ponto, 2017 foi o ano que teve mais queimadas = 78
```

```python
data_por_ano.plot()
plt.suptitle('Quantidade de Área por ano') # Plotagem do resultado acima
```

```python
data['Estado'] = data['Coordenação Regional do ICMBio'].str.extract(r'\b(\w+)$', expand=True) # Para o quarto ponto, foi necessário tratar o campo de coord Reg para obter apenas os Estados,sendo assim, criando uma nova coluna
```

```python
data_by_region =data.groupby(['Estado']).count().sort_values(['2018','2017','2016','2015','2014','2013','2012'],ascending=False) # Agrupando dataframe pelos estados
```

```python
data_by_region['Total Por Estado'] = data_by_region['2018']+data_by_region['2017']+data_by_region['2016']+data_by_region['2015']+data_by_region['2014']+data_by_region['2013']+data_by_region['2012']
#Para obter o total por estado. Criou-se uma nova coluna 'Total por Estado' atribuindo a soma das quantidades de areas queimadas do ano de 2012 a 2018
```

```python
data_by_region.sort_values('Total Por Estado', ascending=False) # Realizando sorte em ordem decrescente por estado
```
