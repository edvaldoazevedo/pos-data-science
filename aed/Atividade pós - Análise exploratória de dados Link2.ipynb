```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('DIMIF_desmatamento.csv', delimiter=';')# Importação do dataset com configuração do delimitador ;
```

```python
dataset = data.iloc[1:,:]#Utilizacao do iloc para remover primeira linha com registro desnecessário
```

```python
dataset.head()
```

```python

des_prodes= dataset[dataset['Monitoramento PRODES'] == 'Sim']# Filtrando apenas os registros que foram monitorados pela PRODES
```

```python
des_prodes.rename(columns={'PRODES 2018':'2018','PRODES 2017':'2017', 'PRODES 2016':'2016','PRODES 2015':'2015','PRODES 2014':'2014'}, inplace = False)
# Renomeando campos para facilitar durante o processo de análise
```

```python
count_by_category_uc=des_prodes.groupby(['Categoria da UC: nomenclatura nacional']) # Agrupando dataframe por categoria da UC
```

```python
des_by_category =count_by_category_uc[['2014','2015','2016','2017','2018']].count() #Pergunta1 : Qual foi a evolulação de desmatamento entre 2014 a 2018 relacionaDO a categoria da UC
```

```python
des_by_category.plot(figsize=(30,10))

plt.suptitle('Desmatamento Por categoria 2014 a 2018') # Apresentação do gráfico para pergunta acima
```

```python
des_prodes['Estado'] = des_prodes['Coordenação Regional do ICMBio'].str.extract(r'\b(\w+)$', expand=True) # Tratamento para filtrar apenas os estados e criação da coluna Estado
```

```python
des_prodes
```

```python
quant_desm_by_region= des_prodes.groupby(['Estado']).count() # Pergunta2 : Qual foi o estado que mais teve desmatamento no ano de 2018
```

```python
quant_desm_by_region[['2018']].sort_values('2018',ascending=False) # Resultado apresentado pela tabela: O Pará foi o estado que mais teve desmatamento no ano de 2018
```

```python
quant_desm_by_categoria= des_prodes.groupby(['Categoria da UC: nomenclatura nacional']).count() # Pergunta3: Qual foi a maior quantidade de desmatamento em 2018 por Categoria de UC
```

```python
quant_desm_by_categoria.sort_values(['2018'],ascending=False)
```
